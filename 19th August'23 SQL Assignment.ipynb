```python
Write an SQL query to report the difference between the number of apples and oranges sold each day.
Return the result table ordered by sale_date.


CREATE TABLE Sales1 (
    sale_date DATE,
    fruit ENUM('Apple', 'Orange', 'Banana', 'Mango', 'Grapes'),
    sold_num INT,
    PRIMARY KEY (sale_date, fruit)
);

INSERT INTO Sales1 (sale_date, fruit, sold_num) VALUES
('2020-05-01', 'Apple', 10),
('2020-05-01', 'Orange', 8),
('2020-05-02', 'Apple', 15),
('2020-05-02', 'Orange', 15),
('2020-05-03', 'Apple', 20),
('2020-05-03', 'Orange', 0),
('2020-05-04', 'Apple', 15),
('2020-05-04', 'Orange', 16);


select sale_date, 
sum(case when fruit='Apple' then sold_num else 0 end) as applesum1,
sum(case when fruit='Orange' then sold_num else 0 end) as orangesum2
from Sales1
 group by sale_date order by sale_date;
```

```python



#Write an SQL query to report the fraction of players that logged in again on the day after the day they
first logged in, rounded to 2 decimal places.



CREATE TABLE Activity1 (
    player_id INT,
    device_id INT,
    event_date DATE,
    games_played INT,
    PRIMARY KEY (player_id, event_date)
);

INSERT INTO Activity1 (player_id, device_id, event_date, games_played)
VALUES
(1, 2, '2016-03-01', 5),
(1, 2, '2016-03-02', 6),
(2, 3, '2017-06-25', 1),
(3, 1, '2016-03-02', 0),
(3, 4, '2018-07-03', 5);


with firstlogin as (
select player_id, min(event_date) as first_login_date from Activity1 group by player_id
)
select  round(count(distinct f1.player_id)/count(distinct a.player_id),2) as fraction_of_player
from firstlogin f1
join Activity1 a on a.player_id=f1.player_id and 
datediff(a.event_date,f1.first_login_date)=1;
```

```python
#Write an SQL query that reports the most experienced employees in each project. In case of a tie,
report all employees with the maximum number of experience years.



INSERT INTO Employee5 (employee_id, name, experience_years)
VALUES
(1, 'Khaled', 3),
(2, 'Ali', 2),
(3, 'John', 3),
(4, 'Doe', 2);

INSERT INTO Project5 (project_id, employee_id)
VALUES
(1, 1),
(1, 2),
(1, 3),
(2, 1),
(2, 4);

select * from Employee5;
select * from Project5;


with max_experience as (select p.project_id, p.employee_id, max(e.experience_years) as max_experience 
from Project5 p join Employee5 e on
p.employee_id=e.employee_id group by project_id)

select p.project_id,e.employee_id,e.name,e.experience_years
from Project5 p
join Employee5 e on p.employee_id=e.employee_id
join max_experience me on p.project_id=me.project_id 
where e.experience_years=me.max_experience;
```
